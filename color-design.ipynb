# 不等距 colorbar

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.colors as mcolors

x = np.linspace(0, 3 * np.pi, 500)
data = np.sin(x - x[:, np.newaxis] * 2) * 95
uneven_levels = [-96, -72, -48, -24, -12, -6, 0, 6, 12, 24, 48, 72, 96]
cmap_rb = plt.get_cmap('RdBu_r')
colors = cmap_rb(np.linspace(0, 1, len(uneven_levels) - 1))
cmap, norm = mcolors.from_levels_and_colors(uneven_levels, colors)

fig, axs = plt.subplots(ncols=2, figsize=(6, 4))

v = np.linspace(-96, 96, 1000)
axs[0].plot(v, norm(v))

cs = axs[1].pcolormesh(data, cmap=cmap, norm=norm)
cbar = fig.colorbar(cs, ticks=uneven_levels)

plt.show()
```

# cmaps拼接

```python
import cmaps
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
state = np.random.RandomState(51423)
data = state.rand(30, 30).cumsum(axis=1)
fig = plt.figure(figsize=(4, 4), dpi=100)
cmap1=cmaps.StepSeq25_r[:10]                                                  
cmap2=cmaps.StepSeq25_r[10:15]                                                 
list_cmap1=cmap1(np.linspace(0,1,10))                                      
list_cmap2=cmap2(np.linspace(0,1,5))                                           
new_color_list=np.vstack((list_cmap1,list_cmap2))                            
new_cmap=ListedColormap(new_color_list,name='new_cmap ')       
plt.pcolormesh(data, cmap=new_cmap)
plt.colorbar()
```

# deep convective cloud

```python
import cmaps
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from matplotlib import colors
mycolor=["yellow","cornflowerblue","royalblue","royalblue"]#"royalblue","cornflowerblue"
cmap_color = colors.LinearSegmentedColormap.from_list('my_list', mycolor)
state = np.random.RandomState(51423)
data = state.rand(30, 30).cumsum(axis=1)
fig = plt.figure(figsize=(17, 8), dpi=100)
cmap1=cmaps.MPL_gist_heat[:80]                                                
cmap2=cmap_color 
cmap3=cmaps.MPL_Greys[:100]
list_cmap1=cmap1(np.linspace(0,1,320))                                      
list_cmap2=cmap2(np.linspace(0,1,320)) 
list_cmap3=cmap3(np.linspace(0,1,400))  
new_color_list=np.vstack((list_cmap1,list_cmap2,list_cmap3))                            
new_cmap=ListedColormap(new_color_list,name='new_cmap ')   
plt.pcolormesh(data, cmap=new_cmap)
plt.colorbar()
```
